Write a Python program to track the remaining balance at the end of each month until the account balance reaches zero.

```python
bank_balance = 5000
month_counter = 0

while bank_balance > 0:
    spending = 1000
    bank_balance -= spending
    month_counter += 1
    print(f'At the end of the month {month_counter}: ' + f'My balance is ${round(bank_balance, 2)}')
```
