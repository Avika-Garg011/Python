### While Loop

1. Multiplication Table:
   
o Write a program that asks the user to input a number and then prints its multiplication table (from 1 to 10) using a while loop.

```python
num = int(input("Enter your number : "))
x = 1
while x <= 10:
    print(f"{num} x {x} = ",x*num)
    x += 1
```

2. Password Validation:
   
o Write a program that asks the user to input a password and continues asking until the password matches a predefined password (e.g., "secret123"). 
If the password is correct, print a success message and break the loop.

```python
correct_password = "secret123"

while True:                                      #This creates an infinite loop that keeps asking for input.
    password = input("Enter password:")
    if password == correct_password:
        print("Password is correct.")
        break
    else:
        print("Password is incorrect. Try again.")
        
```

3. Sum of Numbers Until Condition Met:
   
o Write a program that repeatedly asks the user for a number and adds it to a running total. The loop should continue until the user enters "done". Print the total after the loop finishes.

```python
total = 0 
while True:                                   # When input is required multiple times then it should be inside while 
    num = input("Enter a number or done")
    if num == "done":
        print("The Total value is :",total)
        break                                # Anything after break does not execute 
    else:
        num = int(num)
        total += num
        print("The Total value is :",total)
```

4. Even Numbers:
   
o Write a program that prints even numbers from 1 to 100 using a while loop. If the number is divisible by 4, skip printing that number using the continue statement.

```python
i = 1
while i <= 100:
    if i % 2 == 0 :
        if i % 4 == 0:
            i += 1
            continue
        print(i)
    i += 1      #Go to next iteration 
```

5. Break the Loop on Specific Condition:
   
o Create a program that asks the user to enter numbers. If the user enters the number 0, the loop should break and print "Loop exited".

```python
while True:
    num = int(input("Enter a number: "))

    if num == 0:
        break
    print("You entered num :",num)

print("Loop Exited")
    
```

6. Skip Negative Numbers:
   
o Write a program that asks the user to input numbers. If the input is a negative number, use the continue statement to skip processing that number and move to the next input.

```python
while True:
    num = int(input("Enter any number:"))

    if num < 0:
        print("Negative numbers are not allowed")
        continue                                            #Anything after continue in the loop is skipped.

    print("Number is allowed")
```

9. Count Down Timer:
    
o Write a program that simulates a countdown from a number that the user enters. The program should print the countdown, and when it reaches 0, print
"Time's up!" and stop

```python
num = int(input("Enter the countdown number :"))

while num > 0:
        print(num)
        num -= 1
print("Time's up!")
```

20. Display Numbers (Exit on 'Q'):
    
o Write a program that displays numbers starting from 1. The user can input 'Q' to stop the program. If 'Q' is entered, break the loop.

```python
x = 1

while True:
    print(x)

    choice = input("Press Enter to continue or Q to quit: ")

    if choice.upper() == "Q":
        print("Program Stopped")
        break

    x += 1
    
```
