🎲 Random Number Guessing Game
In this small Python project, we created a fun guessing game using Python’s built-in random library.

🔹 How it works:

The computer generates a random number within a specific range (for example, between 1 and 10).

The player (you) tries to guess what that number is.

After each guess, the game will tell you if your guess is too high, too low, or correct.

The game continues until you guess the correct number.

🔹 What we learned:

✅ How to use the random library to generate random numbers.

✅ How to take input from the user and check conditions using if-else statements.

✅ How to create a simple and interactive Python program that’s fun to play!


```python
import random
```

```python
num = random.randint(1,10)
attempts = 0
guess = None 
print("WELCOME TO THE NUMBER GUESSING GAME!!!")

while guess != num:
    guess = int(input("Your guess: "))
    attempts += 1

    if guess > num:
        print("Too High !!! Guess Again.")
    elif guess < num:
        print("Too Low !!! Guess Again.")
    else:
        print(f"Congrats🎉!!! You guessed it right in {attempts} attempts.")
```
