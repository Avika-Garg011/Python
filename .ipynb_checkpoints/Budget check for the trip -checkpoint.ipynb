In this small Python project, we created a budget-checking program to decide if a trip is possible.

```python
trip = input("Trip? (Yes/No) :").lower()
if trip == "yes":
    print("Trip Budget is Rs. 10,000") 
    budget = int(input("Enter the budget you have : "))
    
    if budget >= 10000:
        print("Trip is on🥳")
    else:
        borrow = input("Do you have want to borrow money ? (Yes/No) : ").lower()
        
        if borrow == "yes": 
            borrow_amount = int(input("How much money can be borrowed?"))
            total = budget + borrow_amount
            
            if total >= 10000:
                print("Trip is on🥳")
            else:
                print("Still not enough money. Trip cancelled.😔")
        else:
            print("No money!!! No Trip😣")
else:
    print("No Trip😔")
```
