🎮 Rock, Paper, Scissors Game
In this small Python project, we created a fun and interactive Rock, Paper, Scissors game using Python’s built-in random library.

🔹 How it works:

The player chooses either rock, paper, or scissors.

The computer randomly selects one of the three options.

The game compares both choices and decides whether it’s a win, loss, or tie for the player.

The player can choose to play again or exit the game.

🔹 What we learned:

✅ How to use the random.choice() function to select a random item from a list.

✅ How to take input from the user and validate it.

✅ How to compare values and decide a winner using if-elif-else statements.

✅ How to loop the game so it can be played multiple times without restarting the program.


This project is a great practice for understanding lists, loops, conditionals, and user input — essential skills for Python programming.

```python
import random

print("🎮 Welcome to Rock, Paper, Scissors!")

choices = ["rock", "paper", "scissors"]

while True:
    user_choice = input("Enter rock, paper, or scissors (or 'exit' to quit): ").lower()

    if user_choice == "exit":
        print("Thanks for playing! 👋")
        break

    if user_choice not in choices:
        print("❌ Invalid choice! Please try again.")
        continue

    computer_choice = random.choice(choices)
    print(f"Computer chose: {computer_choice}")

    if user_choice == computer_choice:
        print("It's a tie! 🤝")
    elif (
        (user_choice == "rock" and computer_choice == "scissors") or
        (user_choice == "paper" and computer_choice == "rock") or
        (user_choice == "scissors" and computer_choice == "paper")
    ):
        print("You win! 🎉")
    else:
        print("You lose! 😢")
```
