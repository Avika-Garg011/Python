1. Write a Python program to calculate the sum of all numbers from 1 to 100 using a for loop.

```python
total = 0 

for number in range(1,101):
    total += number

print("The sum of numbers from 1 to 100 is:", total)
```

2. How do you iterate through a list using a for loop in Python?
   
You can iterate through a list in Python using a for loop by accessing each element one at a time.

```python
fruits = ["Apple", "Banana", "Mango", "Orange"]

for fruit in fruits:
    print(fruit)
```

3. Write a Python program to find the product of all elements in a list using a for loop

```python
numbers = [5,10,2,4]
product = 1 
for elements in numbers:
    product *= elements

print("The product of elements is:", product)
```

4. Create a Python program that prints all even numbers from 1 to 20 using a for loop.

```python
even = []

for i in range(1,21):
    if i%2 == 0:
        even.append(i)

print(even)
```

5. Write a Python program to find the largest number in a list using a for loop.

```python
user_input = input("Enter list of numbers :")
split_values = user_input.split(",")
num = [int(x.strip())
       for x in split_values]
largest = num[0]
for x in num:
    if x > largest:
        largest = x
print("The largest number in the list is:", largest)
```

6. Write a Python program to count the number of vowels in a given string using a for loop.

```python
sentence = input("Enter a string :")
count = 0 
for char in sentence:
    if char.lower() in "aeiou":
        count += 1
print("Number of vowels :",count)
```
